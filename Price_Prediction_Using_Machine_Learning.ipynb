**Task 3 - Data Cleaning**

- Check and remove the duplicate rows

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Convert numerical columns to float
numerical_columns = ['symboling', 'normalized-losses', 'wheel-base', 'length', 'width',
                     'height', 'curb-weight', 'engine-size', 'bore', 'stroke',
                     'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
                     'highway-mpg', 'price']

# Loop through the columns and convert them to float
for col in numerical_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Use 'coerce' to handle non-numeric values
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
df.describe(include=["O"])
```

```python
df.describe()
```

```python
#Imputing missing values
# Columns grouped for imputation
mean_columns = ['wheel-base', 'length', 'width', 'height', 'curb-weight',
                'bore', 'stroke', 'peak-rpm', 'city-mpg', 'highway-mpg']
median_columns = ['symboling', 'normalized-losses', 'engine-size',
                  'compression-ratio', 'horsepower', 'price']

# Apply mean imputation
df[mean_columns] = df[mean_columns].apply(lambda col: col.fillna(col.mean()))

# Apply median imputation
df[median_columns] = df[median_columns].apply(lambda col: col.fillna(col.median()))

# Verify remaining missing values
print(df.isnull().sum())
```

```python
# List of columns to impute
columns_to_impute = ['fuel-type', 'aspiration', 'num-of-doors', 'drive-wheels', 'engine-location','engine-type','num-of-cylinders','fuel-system','body-style']

# Impute each column with its mode (most frequent value)
for col in columns_to_impute:
    mode_value = df[col].mode()[0]  # Get the most frequent value (mode)
    df[col] = df[col].fillna(mode_value)  # Fill missing values with the mode
```

```python
df.isnull().sum()
```

```python
# Replace '?' with NaN in the 'num-of-doors' column
df['num-of-doors'] = df['num-of-doors'].replace('?', np.nan)

# Get the mode (most frequent value) of the 'num-of-doors' column
mode_value = df['num-of-doors'].mode()[0]

# Impute missing values with the mode
df['num-of-doors'] = df['num-of-doors'].fillna(mode_value)
df['num-of-doors'].value_counts()
```

```python
# Group by the columns 'fuel-type', 'aspiration', and 'num-of-doors'and impute the 'make' column with the mode (most frequent value) within each group
df['make'] = df.groupby(['fuel-type', 'aspiration', 'num-of-doors'])['make'].transform(lambda x: x.fillna(x.mode()[0]))
```

Explanation:
groupby(['fuel-type', 'aspiration', 'num-of-doors']): This groups the DataFrame by the specified columns.
['make'].transform(lambda x: x.fillna(x.mode()[0])): For each group, the most frequent value (mode()) is calculated, and the missing values (NaN) in the make column are filled with this mode value. The transform() function ensures that this operation is applied back to the entire column.
This code will fill the missing make values with the most frequent make value for each unique combination of fuel-type, aspiration, and num-of-doors.
